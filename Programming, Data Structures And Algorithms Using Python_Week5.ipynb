```python
l = list()
d_stu = dict()
d_cgp = {'A':10,'AB':9,'B':8,'BC':7,'C':6,'CD':5,'D':4}
d_gra = dict()

# p = ""           
# while(True):
#     p = input()
#     l.append(p)
#     if p == 'EndOfInput':
#         break

fhand = open(r'D:\Python_Tutorials\aa.txt')                    # code for testing 
for line in fhand:                                             # taking the entire text as a file 
    line = line.rstrip()
    l.append(line)


courses = l[1:l.index('Students')]                               # separating text into sections
students = l[l.index('Students')+1:l.index('Grades')]
grades = l[l.index('Grades')+1:l.index('EndOfInput')]
#print(l)
#print(courses)
#print(students)
#print(grades)

grades_c = list(range(len(grades)))                          # creating a list of list containing separated grade text
for i in range(len(grades)):
    grades_c[i] = grades[i].split('~')

#print('grades_c',grades_c)

for i in range(len(students)):                                       # creating a dict for name of each roll no
    d_stu[students[i][0:7]] = students[i][8:]

d_stu = dict(sorted(d_stu.items()))                                #sorting the dict according to roll no which is key here
#print('d_stu',d_stu)

# for k in d_stu:
#       print(k,d_stu[k])

for k in d_stu:                                              #creating a dict for total cgp for each roll no
    for j in range(len(grades_c)):  
        if k in grades_c[j]:
            if k not in d_gra:
                d_gra[k] = d_cgp[grades_c[j][-1]]
            else:
                d_gra[k] = d_gra[k] + d_cgp[grades_c[j][-1]]            
            
#print('d_gra',d_gra)    

l2 = list()                                  ##breaking the grade text into making individual words as elements of list
for i in range(len(grades_c)):
    l2 = l2 + grades_c[i]
#print('l2',l2) 

d_count = dict()                      #creating a dict for counting no of subjects for which grade is recorded of each rollno
for j in d_stu:
    d_count[j] = l2.count(j)
#print('d3',d3) 

d_avg = dict()                                              #creating a dict for avg cgp for each roll no
for j in d_stu:
    if d_count[j] != 0:
        d_avg[j] = round(d_gra[j]/d_count[j],2)
    else:
        d_avg[j] = 0
#print('d_avg',d_avg)
           
            
for i in d_avg:
    print(str(i)+'~'+d_stu[i]+'~'+str(d_avg[i]))
```
